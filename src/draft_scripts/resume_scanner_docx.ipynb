```python
def preprocess(document):
        '''
        Information Extraction: Preprocess a document with the necessary POS tagging.
        Returns three lists, one with tokens, one with POS tagged lines, one with POS tagged sentences.
        Modules required: nltk
        '''
        try:
            # Try to get rid of special characters
#             try:
#                 document = document.decode('ascii', 'ignore')
#             except:
#                 document = document.encode('ascii', 'ignore')
            # Newlines are one element of structure in the data
            # Helps limit the context and breaks up the data as is intended in resumes - i.e., into points
            lines = [el.strip() for el in document.split("\n") if len(el) > 0]  # Splitting on the basis of newlines 
            lines = [nltk.word_tokenize(el) for el in lines]    # Tokenize the individual lines
            lines = [nltk.pos_tag(el) for el in lines]  # Tag them
            # Below approach is slightly different because it splits sentences not just on the basis of newlines, but also full stops 
            # - (barring abbreviations etc.)
            # But it fails miserably at predicting names, so currently using it only for tokenization of the whole document
            sentences = nltk.sent_tokenize(document)    # Split/Tokenize into sentences (List of strings)
            sentences = [nltk.word_tokenize(sent) for sent in sentences]    # Split/Tokenize sentences into words (List of lists of strings)
            tokens = sentences
            sentences = [nltk.pos_tag(sent) for sent in sentences]    # Tag the tokens - list of lists of tuples - each tuple is (<word>, <tag>)
            # Next 4 lines convert tokens from a list of list of strings to a list of strings; basically stitches them together
            dummy = []
            for el in tokens:
                dummy += el
            tokens = dummy
            # tokens - words extracted from the doc, lines - split only based on newlines (may have more than one sentence)
            # sentences - split on the basis of rules of grammar
            return tokens, lines, sentences
        except Exception as e:
            print(e) 
```

```python
# Read docx files 
# import os
#from docx import Document

def readFile(fileName):
        '''
        Reads docx resume given its name as a string.
        Modules required: os, Document (from docx)
        packages required: python-docx
        Return:
        output_1: resume line by line
        output_2: resume plain text
        '''
        extension = fileName.split(".")[-1]
        
        if extension == "docx":
            try:
                document = Document(fileName)
                output_1 = "\n".join([pgraph.text for pgraph in document.paragraphs]) # By line mode
                output_1 = output_1.split("\n")
                output_2 = ' '.join([pgraph.text for pgraph in document.paragraphs]) # Plain text mode
                return output_1, output_2, extension
            
            except:
                print('####### WARNING ####### WARNING ####### WARNING #######')
                print(fileName+' ========> IS EMPTY ')
                print('####### WARNING ####### WARNING ####### WARNING #######')
                return 'NO_DATA', 'NO_DATA', extension
                pass
        
#         elif extension == "pdf":
#             pass
#             try:
#                 return convertPDFToText(fileName), extension
#             except:
#                 return ''
#                 pass
        else:
            print('Unsupported format')
            return '', '', ''
```

```python
#import glob
## READ DOCX FILES

# Determine how many resume files are there
docx_files = glob.glob("../../data/resumes/*.docx")
pdf_files = glob.glob("../../data/resumes/*.pdf")

files = set(docx_files + pdf_files)
files_list = list(files)
print("%d files identified" %len(files))
print("%d docx files " %len(docx_files))
print("%d pdf files" %len(pdf_files))

info_dict = {}
files_missing_unique_id = []
files_missing_store_id = []

count = 0 # Count initialization

print("\n############ Processing docx files #############\n")

file_dir = "../../data/resumes/"

# Read docx files
for root, dirs, files in os.walk(file_dir): #This also list files inside sub directories
    for file in files:
        
        file_extension = file.split(".")[-1]
        
        if file_extension in ["doc","rtf", "pdf", "txt", "xlsx"]: #Skip other file types
            skip=True
            continue
        if file.startswith('~'): #Ignore hidden files
            skip=True
            continue
        else:
            file_name = file.split(".")[0]
            #print(file_name)
            file_path_name = os.path.join(root, file)
            #print(file_path_name)
            
# Read Resume data and extension
        count += 1 
        print("Reading file %s"% count +" " +file)
        
        resume_bline, resume_text, file_type = readFile(file_path_name)   # Call read function
        
# Read employee name, store_id and unique id from meta data
        splits = file_name.split(" - ")
        employee_name = splits[0]
        if len(splits) != 1:
            store_id = splits[1]
            if len(splits) == 3:
                unique_id = splits[2]
            else: 
                unique_id = 'NO_UID'
                #unique_id = ''
                files_missing_unique_id.append(file)
                print(employee_name+' ========> DOES NOT HAVE UNIQUE ID')
        else: 
            store_id = 'NO_SID'
            #store_id = ''
            files_missing_store_id.append(file)
            print(employee_name+' ========> DOES NOT STORE ID')

# Save properties to dictionary
        info_dict.setdefault("file_name",[]).append(file_name)
        info_dict.setdefault("file_type",[]).append(file_type)
        info_dict.setdefault("meta_e_name",[]).append(employee_name)
        info_dict.setdefault("meta_s_id",[]).append(store_id)
        info_dict.setdefault("meta_u_id",[]).append(unique_id)
        info_dict.setdefault("resume_bline",[]).append(resume_bline) #Show imported resume by line
        info_dict.setdefault("resume_text",[]).append(resume_text) #Show imported resume by line

# Create missing uid and sid files
file_path_name = os.path.join(file_dir, 'docx_files_missing_meta_unique_id.txt')
print(file_path_name)
f = open(file_path_name,'w')
f.write("Total of {} employees are missing unique_ids\n".format(len(files_missing_unique_id)))
for item in files_missing_unique_id:
    f.write(item+'\n')
f.close() 

file_path_name = os.path.join(file_dir, 'docx_files_missing_meta_store_id.txt')
f = open(file_path_name,'w')
f.write("Total of {} employees are missing store_ids\n".format(len(files_missing_store_id)))
for item in files_missing_store_id:
    f.write(item+'\n')
f.close() 

print("\nTotal of {} employees are missing unique_ids".format(len(files_missing_unique_id)))
print("Total of {} employees are missing store_ids\n".format(len(files_missing_store_id)))

print("employees missing unique ids saved in `files_missing_meta_unique_id.txt`.")
print("employees missing store ids saved in `files_missing_meta_store_id.txt`.")
            
#Convert to pandas dataframe    
docx_df = pd.DataFrame(info_dict)
print("\nSave results into docx_df pandas dataframe variable")

# Export as csv file to the output folder
ouput_path_name = "../../data/output/docx_english_resume.csv"
docx_df.to_csv(ouput_path_name)
print("Writing CSV file to {} \n".format(ouput_path_name))

print("Done!")
```
